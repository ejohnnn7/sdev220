Binary Search to Find an Element

Given a sorted array and an integer k, the task is to find the index where k appears using binary search.

- If k is not present, return -1
- If k appears multiple times, return the first occurrence

```python
def binary_search_first(arr, k):
    # left pointer
    left = 0
    # right pointer
    right = len(arr) - 1

    # store the result index
    result = -1

    # continue searching while the range is valid
    while left <= right:
        # find middle index
        mid = (left + right) // 2

        # if we find k
        if arr[mid] == k:
            # save this index
            result = mid 
            # keep searching for the first occurrence
            right = mid - 1

        # if k is larger, ignore left half
        elif arr[mid] < k:
            left = mid + 1

        # if k is smaller, ignore right half
        else:
            right = mid - 1

    return result
```

```python
# test cases
print(binary_search_first([1, 2, 3, 4, 5], 4))        # expected: 3
print(binary_search_first([11, 22, 33, 44, 55], 445)) # expected: -1
print(binary_search_first([1, 1, 1, 1, 2], 1))        # expected: 0
```
