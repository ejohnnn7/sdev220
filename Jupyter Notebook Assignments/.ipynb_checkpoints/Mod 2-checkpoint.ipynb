```python
# choose a number between 1 and 10 and assign to variable "secret"
secret = 7

# choose a number between 1 and 10 and assign variable "guess"
guess = 2

# conditional checks
if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

4.2

```python
# assign true or false values to the variables "small" and "green"
small = True
green = False

# Cherry
if small and not green:
    print("Cherry")

# Pea
if small and green:
    print("Pea")

# Watermelon
if not small and green:
    print("Watermelon")

# Pumpkin
if not small and not green:
    print("Pumpkin")
```

6.2

```python
# assign the value 7 to the variable "guess_me"
guess_me = 7

# assign the value 1 to the variable "number"
number = 1

# conditional checks
while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
    else:
        print("oops")
        break

    number += 1
```

```python
# assign the value 5 to the variable "guess_me"
guess_me = 5

# conditional checks
for number in range (10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
