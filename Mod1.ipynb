3.1

There are 60 seconds in a minute, and 60 minutes in an hour. To find the number of seconds in an hour, multiply 60 by 60.

```python
60 * 60
```

3.2

The result from the previous calculation is stored in a variable named 'seconds_per_hour'.

```python
seconds_per_hour = 3600
seconds_per_hour
```

3.3

There are 24 hours in a day. Using the 'seconds_per_hour' variable, I can calculate the total number of seconds in a day.

```python
seconds_per_hour * 24
```

3.4

The total number of seconds in a day is stored in a variable named 'seconds_per_day'.

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

3.5

Floating-point division is performed using the '/' operator.

```python
seconds_per_day / seconds_per_hour
```

3.6

Integer Division is performed using the '//' operator. The result is the same as the floating-point division result, except that the integer does not include a decimal point.

```python
seconds_per_day // seconds_per_hour
```
